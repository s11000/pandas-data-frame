```python
def binary_search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')

if __name__ == '__main__':
    data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    print(binary_search(data, 8))
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def run_algo(n, func):
    start_time = time.time()  # current time
    func(n)
    end_time = time.time()
    return end_time - start_time

def linear(n):
    for i in range(n):
        i

def quadratic(n):
    for i in range(n):
        for j in range(n):
            j

def quadratic2(n):
    for i in range(n):
        for j in range(i):
            j

def cubic(n):
    for i in range(n):
        for j in range(n):
            for k in range(n):
                k

def log_n(n):
    low = 1
    high = n
    while low <= high:
        mid = (low + high) // 2
        if mid * mid == n:
            return mid
        elif mid * mid < n:
            low = mid + 1
        else:
            high = mid - 1

    return -1

numberOfPoints = [60, 60, 60, 60, 1000]
min_value = [500, 50, 50, 10, 10**50]
max_value = [100000, 1000, 1000, 100, 10**101]
list_of_functions_to_run = [linear, quadratic, quadratic2, cubic, log_n]
titles = ["O(n)", "O(n^2)", "O(n^2)", "O(n^3)", "O(log n)"]
degrees = [1, 2, 2, 3, 1]

for i, func in enumerate(list_of_functions_to_run):
    time_complexity = []
    n_values = np.linspace(min_value[i], max_value[i], numberOfPoints[i])
    for n in n_values:
        time_complexity.append(run_algo(int(n), func))
    # Smooth out the time complexity curve using a moving average
    window_size = 4
    moving_average = np.convolve(time_complexity, np.ones(window_size)/window_size, mode='valid')
    # Plot the time complexity curve and the smoothed curve
    plt.figure()
    plt.plot(n_values[len(n_values)-len(moving_average):], moving_average, label='Smoothed')
    # Find a polynomial fit
    x_values_for_fit = n_values[len(n_values)-len(moving_average):]
    label='Polynomial Fit'
    if titles[i] == "O(log n)":
        x_values_for_fit = np.log(x_values_for_fit)
        label='Logarithmic Fit'
    polyfit = np.polyfit(x_values_for_fit, moving_average, deg=degrees[i])
    polyfit_fn = np.poly1d(polyfit)
    plt.plot(n_values, polyfit_fn(n_values), label=label,color='tab:blue', linestyle='--')
    # Set the title and axis labels
    plt.title(f"Time Complexity: {titles[i]}")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Time Complexity")
    plt.legend()

# Show the plot
plt.show()
```
