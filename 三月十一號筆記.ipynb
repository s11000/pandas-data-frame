# 5.Universal function 與向量化運算(Vectorization computation)

問題:計算0的三次方到999的三次方

解法大PK:

Python ==> For loop

```python
 ar=range(1000)
%timeit [ar[i]**3 for i in ar]
ar
```

```python
for i in range (0,1000):
    ar = i**3
    print(i,ar)
    
 
 
```

```python
for i in [0,1,2,3,4,5]:
    square = i**999
    print(i,'squared is',square)
     
 
print('The for loop is complete!')
```
