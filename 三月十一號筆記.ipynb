## ndarray的屬性:軸(axis)

```python
import numpy as np
np.sum([[0, 4, 2], [-2, 5, 3]], axis=0)
```

```python
np.arange(12).reshape(4, -1).shape
```

```python
np.arange(16).reshape(4, -1).shape
```

```python
np.arange(20).reshape(4, -1).shape
```

```python
import numpy as np
np.array = numpy.arange(1,11).reshape(2,5)

print(np.array)
```

```python
import numpy as np
np.array = numpy.arange(1,10).reshape(3,3)

print(np.array)
```

```python
import numpy as np
np.array = numpy.arange(1,10).reshape(3,3)

print(np.array)
```
